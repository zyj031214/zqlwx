```python
# 单变量描述统计

## 无序类别变量描述统计

import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"]# 设置字体
df=pd.read_spss(R"C:\Users\34426\Desktop\zqlwx\identity.sav")
```

```python
df["民族"].value_counts(normalize=True)
```

```python
x = df["民族"].value_counts().index
y = df['民族'].value_counts(normalize=True).values * 100
fig, ax2 = plt.subplots()
labels=df["民族"].value_counts().index
ax2.pie(y, autopct="%1.1f%%", labels=labels,textprops=dict(color="w"))
plt.show()
```

```python
# 有序定类变量描述统计，可增加累计次数和累计比例的统计。可采用饼图或柱状图进行可视化
df["年级"].value_counts()
```

```python
from pyreadstat import pyreadstat
df2, metadata = pyreadstat.read_sav(R'identity.sav',apply_value_formats=True,formats_as_ordered_category=True)
result = df2['年级'].value_counts(sort=False)
```

```python
df2["年级"].value_counts(sort=False)
```

```python

result = pd.DataFrame(df2['年级'].value_counts(sort=False))
df2_result = (result)
df2_result['累计求和'] = result.values.cumsum()
#df2_result['新变量1']=1
#df2_result['新变量2']=df2_result.新变量1+1
df2_result['比例']=df2['年级'].value_counts(normalize=True)
df2_result['累计比例']=df2_result['比例'].cumsum()
```

```python
x=df2_result.index
y=df2_result['count'].values
fig, ax2 = plt.subplots()
ax2.scatter(x,y)
plt.show()
```
